**Analyzing the iris dataset and predict the optimum number of clusters and
represent it visually.**

```
#importing the important libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```
#Loading the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head(10) # getting first 10 rows
```

**Finding the number of clusters or optimum value of k**

```
x = iris_df.iloc[:, [0, 1, 2, 3]].values
wcss=[]
#Standard code for implementing K-means clustering upto  value 10
for j in range(1,11):
  kmeans = KMeans(n_clusters = j, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0) 
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

**Plotting the results onto a line graph, allowing us to observe 'The elbow'**

```
plt.figure(figsize=(16,8))
plt.plot(range(1, 11), wcss,'y')
plt.title('The elbow method')
plt.xlabel('------------------------Number of clusters------------------------')
plt.ylabel('-------------------------WCSS--------------------------------------')
plt.show()
```

**From the above shown graph we can easliy determine that at value k=3,we can observe Albow and also this is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.**

**Applying K-means clustering and considering value of k=3 in the dataset**

```
# Applying the K-means to the ris  dataset and Creating the Kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

**Predicting the K-means classifcation using the graph representation**

```
plt.figure(figsize=(14,6))
plt.scatter(x[:,0],x[:,1],c=y_kmeans,s=50,cmap='rainbow')
plt.title('Graph representing prediction using Sepal Length and Width ')
plt.xlabel('------------Sepal Length-----------')
plt.ylabel('------------Sepal Width-----------')
```

```
plt.figure(figsize=(14,6))
plt.scatter(x[:,2],x[:,3],c=y_kmeans,s=50,cmap='rainbow')
plt.title('Graph representing prediction using Petal Length and Width ')
plt.xlabel('------------Petal Length-----------')
plt.ylabel('------------Petal Width-----------')
```
